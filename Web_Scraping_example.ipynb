```python
from pprint import pprint
from urllib import request
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import pymysql
```

```python
#Get top 25 teams for 2017 Week 10
college_hoops_ranking = "http://www.espn.com/mens-college-basketball/rankings/_/year/2017/week/10/seasontype/2"

test_url = request.urlopen(college_hoops_ranking)

read_html = test_url.read()
```

```python
soup = BeautifulSoup(read_html)
print(soup.prettify())
```

```python
getdata = soup.html.find_all("div", class_="mod-content")

type(getdata)
```

```python
getdata

```

```python
rankings = getdata[1]
```

```python
rankings
```

```python
rows = rankings.findChildren(['td'])
rows
```

```python
x = 0
end = len(rows)

results = []

while x < end:
    for y in rows[x]:
        if y.string == None:
            for link in rankings.find_all('a'):
                results.append(link.get('href'))
        else:
            results.append(y.string.split(' '))
            
    x = x + 1

results
```

```python
final = []

for x in results:
    if len(x) < 3:
        final.append(x)
        
final
```

```python
data = []

a = 4
b = a + 4
end = len(final)

for x in final:
    while a < end:
        data.append(final[a:b])
        
        a = a + 4
        b = a + 4

data
```

```python
df = pd.DataFrame({
        'RK':[x[0] for x in data],
        'TEAM':[x[1] for x in data],
        'RECORD':[x[2] for x in data],
        'PTS':[x[3] for x in data],
        'WEEK': 10
    })

df
```

```python
#Remove brackets from data values
df['TEAM'] = df['TEAM'].str[0] + ' ' +df['TEAM'].str[1].fillna('') 
df['RK'] = df['RK'].str[0]
df['RECORD'] = df['RECORD'].str[0]
df['PTS'] = df['PTS'].str[0]

#Remove comma from PTS column to ensure it can become an intger
df['PTS'] = df['PTS'].str.replace(',','')

#Convert string to int for PTS column
df['PTS'] = df['PTS'].astype(int)

df
```
